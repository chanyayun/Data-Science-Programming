```python
from selenium import webdriver
from pyquery import PyQuery as pq
```

```python
driver = webdriver.Chrome("../../../../pytest/chromedriver.exe")
```

```python
import time
```

```python
def fetch_news(news_url):

    #news tab
    driver.get(news_url)
    html = driver.find_element_by_css_selector('*').get_attribute("outerHTML")
    doc = pq(html)

    title = doc('article > div.l-container > h1').text()
    date = doc('body article > div.l-container > div.metadata > div.metadata__info.js-byline-images > p.update-time').text()
    content = doc('#body-text > div.l-container > div.zn-body__read-all').text()

    print(date)
#     print(date.split(") ")[1])
    print(title)
    
    #merge text
    full_news = [title, content]
    text = " ".join(full_news)
    
    #close new tab and switch back
    driver.close()
    driver.switch_to.window(driver.window_handles[0])
    
    #check year
    if "2019" not in date:
        print(date + "--> from fetch_news(): no more 2019")
        boo_check = False
    else:
        text_list.append(text)
        boo_check = True
```

```python
def fetch_travel(news_url):

    driver.get(news_url)
    html = driver.find_element_by_css_selector('*').get_attribute("outerHTML")
    doc = pq(html)

    title = doc('div.Zone__component.Article__zone > div.Article__pageTop > div.Article__wrapper.Article__topWrapper > div.Article__pageHead.Article__pad > h1').text()
    date = doc('div.Zone__component.Article__zone > div.Article__pageTop > div.Article__wrapper.Article__topWrapper > div.Article__pageHead.Article__pad > div').text()
    content = doc('div.Zone__component.Article__zone > div:nth-child(2) > div.Article__primary > div > div > div.Article__pad > div').text()

    #updat'ed', publish'ed'
#     date = date.split("ed ")[1]
    print(date)
    print(title)
    
    #merge text
    full_news = [title, content]
    text = " ".join(full_news)
    
    #close new tab and switch back
    driver.close()
    driver.switch_to.window(driver.window_handles[0])
    
    #check year
    if "2019" not in date:
        print(date + "--> from fetch_travel(): no more 2019")
        boo_check = False
    else:
        text_list.append(text)
        boo_check = True
```

```python
def fetch_style(news_url):
    
    #news tab
    driver.get(news_url)
    html = driver.find_element_by_css_selector('*').get_attribute("outerHTML")
    doc = pq(html)

    title = doc('#mount > div > div.Chrome__content > div > div > div.Zone__component.BasicArticle__zone.BasicArticle__head > div > div > h1').text()
    date = doc('#mount > div > div.Chrome__content > div > div > div.Zone__component.BasicArticle__zone.BasicArticle__head > div > div > div.PageHead__published').text()
    content = doc('#mount > div > div.Chrome__content > div > div > div:nth-child(4) > div > div > div.BasicArticle__bodyTop > div.BasicArticle__main').text()


#     date = date.split("Updated ")[1]
    print(date)
    print(title)
    
    #merge text
    full_news = [title, content]
    text = " ".join(full_news)
    
    #close new tab and switch back
    driver.close()
    driver.switch_to.window(driver.window_handles[0])
    
    #check year
    if "2019" not in date:
        print(date + "--> from fetch_style(): no more 2019")
        boo_check = False
    else:
        text_list.append(text)
        boo_check = True
```

```python
#cnn world 裡面還有 live news ^_^
def fetch_live(news_url):
    
    #news tab
    driver.get(news_url)
    html = driver.find_element_by_css_selector('*').get_attribute("outerHTML")
    doc = pq(html)

    title = doc('#mount > div.Box-sc-1fet97o-0.bRFjwV > article > div.f56c47ff.undefined > div > div._56b5b25c > div.headline > h1').text()
    date = doc('#mount > div.Box-sc-1fet97o-0.bRFjwV > article > div.f56c47ff.undefined > div > div._56b5b25c > div.headline > div > div').text()
    content = doc('#mount > div.Box-sc-1fet97o-0.bRFjwV > article > div._7eda40eb._34e5cb16._4564d500').text()

#     date = date.split("Updated ")[1]
    print(date)
    print(title)
    
    #merge text
    full_news = [title, content]
    text = " ".join(full_news)
    
    #close new tab and switch back
    driver.close()
    driver.switch_to.window(driver.window_handles[0])
    
    #check year
    if "2019" not in date:
        print(date + "--> from fetch_live(): no more 2019")
        boo_check = False
    else:
        text_list.append(text)
        boo_check = True
```

```python
def fetch(news_url):
    if "cnn.com/travel" in news_url:
        fetch_travel(news_url)
    elif "cnn.com/style" in news_url:
        fetch_style(news_url)
    elif "live-news" in news_url:
        fetch_live(news_url)
    else:
        fetch_news(news_url)
```
